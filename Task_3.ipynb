```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data=pd.read_csv('C:\\Users\\vardhan\\Downloads\\householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])

#adding title to the plot
plt.title("Scatter Plot")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#line chart with year against own
plt.plot(data['year'])
plt.plot(data['own'])

#adding title to the plot
plt.title("Line chart")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#bar chart or bar plot
plt.bar(data['year'],data['own'])

#adding title to the plot
plt.title("Bar Chart")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

plt.show()
```

```python
#histogram
plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```
